Does going to university in a different country affect your mental health? A Japanese international university surveyed its students in 2018 and published a study the following year that was approved by several ethical and regulatory boards.

The study found that international students have a higher risk of mental health difficulties than the general population, and that social connectedness (belonging to a social group) and acculturative stress (stress associated with joining a new culture) are predictive of depression.


Using the `students` data and by using PostgreSQL I will be finding out if I would come to a similar conclusion for international students and see if the length of stay is a contributing factor.

Here is a data description of a few columns:

| Field Name    | Description                                      |
| ------------- | ------------------------------------------------ |
| `inter_dom`     | Types of students (international or domestic)   |
| `japanese_cate` | Japanese language proficiency                    |
| `english_cate`  | English language proficiency                     |
| `academic`      | Current academic level (undergraduate or graduate) |
| `age`           | Current age of student                           |
| `stay`          | Current length of stay in years                  |
| `todep`         | Total score of depression (PHQ-9 test)           |
| `tosc`          | Total score of social connectedness (SCS test)   |
| `toas`          | Total score of acculturative stress (ASISS test) |

```r
-- Running this code to save the CSV file as students
SELECT * 
FROM 'students.csv';
```

```r
-- Finding the number of international students and their average scores by length of stay, in descending order of length of stay
SELECT stay, 
       COUNT(*) AS count_int,
       ROUND(AVG(todep), 2) AS average_phq, 
       ROUND(AVG(tosc), 2) AS average_scs, 
       ROUND(AVG(toas), 2) AS average_as
FROM students
WHERE inter_dom = 'Inter'
GROUP BY stay
ORDER BY stay DESC;
```

**Interpretation:**

- **Risk of Depression:** The study shows that how long international students have been in a country doesn't always predict how likely they are to feel depressed. For example, students who've been in the country for 10 years had higher depression scores. This means that feeling down can depend a lot on what each student goes through, not just how long they've been away from home.

- **Social Connectedness:** How connected students feel to others doesn't always match up with how long they've been in the country. Even students who've been around for a while, like 5 years, might not feel very close to others. This tells us that just being in a place for a long time doesn't automatically make you feel like you belong.

- **Acculturative Stress:** Students who've been in the country for a middle amount of time, between 4 and 6 years, often feel more stress from trying to fit into the new culture compared to those who've stayed shorter or longer. This could mean that the stress of adjusting hits hardest during these years but might lessen as students get more used to their new surroundings.

**Conclusion:**

The study finds that international students definitely face challenges with their mental health, but these challenges don't just go up or down based on how long they've been in the country. Other things, like how well they can make friends or deal with changes, are also really important. The average scores of acculturative stress gradually increase with the duration of stay, reaching their peak at the 8-year mark.This matches with the output found out about international students being more at risk for mental health issues, and how feeling connected and dealing with cultural changes can affect their mood.
