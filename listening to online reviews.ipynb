## Review Length and Sentiment Analysis

### Plotting a histogram of review length (in characters) and calculating the mean. Then finding out what is considered a long review and what is considered a short review.


```r
# Calculating the Review length
reviews$review_length <- nchar(as.character(reviews$review_text))
# Calculating the mean of Review Length and defining the long review threshold y using the mean.
mean_review_length <- mean(reviews$review_length)
mean_review_length
long_review_threshold <- mean_review_length 

# Classifying reviews as 'Long' or 'Short' based on the threshold
reviews$review_type <- ifelse(reviews$review_length > long_review_threshold, "Long", "Short")

# Create a histogram of review lengths
ggplot(reviews, aes(x=review_length)) +
  geom_histogram(bins=30, fill="orange", color="black") +
  ggtitle("Histogram of Review Lengths") +
  xlab("Review Length (in Characters)") +
  ylab("Frequency") +
  theme_minimal()

```

### Histogram of Review Length

The histogram shows a right-skewed distribution of review lengths, with a significant concentration of reviews having fewer characters and a tail extending towards longer reviews. The mean review length is 404.13 characters. We define a "long" review as one that exceeds one standard deviation above the mean, and a "short" review as one that falls below this threshold. This operational definition will help categorize reviews quantitatively for further sentiment analysis.

### VADER Sentiment Analysis


```r
# Adding a column to classify reviews as 'Long' or 'Short'
reviews$LengthType <- ifelse(nchar(reviews$review_text) > mean_review_length, "Long", "Short")
```

```r
install.packages('sqldf')
library(sqldf)
install.packages("vader")
library(vader)
# calculating VADER sentiments
# creating a placeholder table, "d"
d <- sqldf("
  SELECT  review_id AS ReviewId, review_text AS ReviewText, 
          0 AS Pos, 0 AS Neu, 0 AS Neg, 0 AS But, 0 AS Compound
  FROM  reviews
")

# then, loop through each review and calculate VADER
for (i in 1:nrow(d)){ #will take ~2 to 10 min., literally
#for (i in 1:5){
  voutput <- get_vader(d[i,2])
  pos <- as.numeric(voutput["pos"])
  neu <- as.numeric(voutput["neu"])
  neg <- as.numeric(voutput["neg"])
  but <- as.numeric(voutput["but_count"])
  compound <- as.numeric(voutput["compound"])
  
  # store the VADER scores inside the placeholder table "d"
  d[i,3] <- pos
  d[i,4] <- neu
  d[i,5] <- neg
  d[i,6] <- but
  d[i,7] <- compound
}
```

```r
# a sql INNER JOIN to merge reviews with the VADER sentiments we calculated
reviews <- sqldf(
  "SELECT r.*, d.pos, d.neu, d.neg, d.but, d.compound
   FROM   reviews r INNER JOIN d ON
    r.review_id = d.ReviewId
  ")
```

- ### Finding if Sentiment Scores are statisitcally different for 1 star Reviews , LONG VS SHORT 

```r
# Performing t-test for Compound sentiment scores of 1-star reviews between Long and Short reviews
t_test_result <- t.test(
  reviews[reviews$rating == 1 & reviews$LengthType == "Long", ]$Compound,
  reviews[reviews$rating == 1 & reviews$LengthType == "Short", ]$Compound
)

t_test_result
```

The Welch Two Sample t-test yielded a statistically significant result (t = 3.1681, df = 72.02, p-value = 0.002252), indicating a notable difference in the mean compound sentiment scores between long and short 1-star reviews. The 95 percent confidence interval (0.1180042 to 0.5185319) further supports this finding, suggesting that the true difference in means is not equal to zero. Specifically, the mean compound sentiment score for long 1-star reviews (0.40673913) is substantially higher compared to short 1-star reviews (0.08847107), suggesting that long 1-star reviews tend to convey stronger overall sentiments, be it positive or negative, compared to their shorter counterparts.

# VADER vs. Positive-to-Negative Ratio


## VADER vs. Positive-to-Negative Ratio
Sentiment analysis is crucial for understanding customer feedback and gauging overall sentiment. Two popular methods for this are VADER (Valence Aware Dictionary and sEntiment Reasoner) and the positive-to-negative ratio method. VADER is a lexicon and rule-based sentiment analysis tool specifically designed for social media texts. It scores text on a scale from -1 (most negative) to +1 (most positive) by considering the intensity and context of words, providing a nuanced sentiment score. On the other hand, the positive-to-negative ratio method is simpler and more direct: it counts the number of positive reviews (e.g., 4- and 5-star ratings) and divides by the number of negative reviews (e.g., 1- and 2-star ratings). This ratio offers a straightforward metric for overall sentiment, where a higher ratio indicates more positive sentiment. While VADER captures subtle emotional tones and provides a detailed analysis, the positive-to-negative ratio method offers a clear, easy-to-understand snapshot of customer satisfaction.


### Identifying the product with the highest average VADER scores:

```r
library(dplyr)

# Calculating the average VADER compound score for each product
average_vader_scores <- reviews %>%
  group_by(product_id) %>%
  summarise(avg_compound = mean(compound, na.rm = TRUE)) %>%
  arrange(desc(avg_compound))

# Displaying the product with the highest average VADER score
highest_vader_product <- average_vader_scores[1, ]
highest_vader_product

## P387511 has the highest average Compound VADER Score of 0.88
```

### Identifying the product with the highest positive-to-negative ratio:


```r
# Categorizing reviews based on star ratings
reviews <- reviews %>%
  mutate(rating_category = case_when(
    rating %in% c(4, 5) ~ "positive",
    rating %in% c(1, 2) ~ "negative",
    TRUE ~ "neutral"
  ))

# Calculating positive-to-negative ratio for each product
positive_negative_ratios <- reviews %>%
  filter(rating_category %in% c("positive", "negative")) %>%
  group_by(product_id) %>%
  summarise(
    positive_count = sum(rating_category == "positive"),
    negative_count = sum(rating_category == "negative"),
    pos_neg_ratio = ifelse(negative_count == 0, Inf, positive_count / negative_count)
  ) %>%
  arrange(desc(pos_neg_ratio))

# Displaying the product with the highest positive-to-negative ratio
highest_ratio_product <- positive_negative_ratios[1, ]
highest_ratio_product

#P416341 is the product witht the highest Positive to negative ratio 
```


- **VADER Sentiment Analysis**: The product with the highest average VADER score is `P387511` with an average compound score of 0.88, indicating very positive sentiment.
- **Positive-to-Negative Ratio**: The product with the highest positive-to-negative ratio is `P416341` with a ratio of approximately 4.09, showing a high number of positive reviews compared to negative reviews.

### Conclusion
In summary, VADER provides a nuanced and detailed sentiment analysis, capturing subtle emotional tones in reviews. The positive-to-negative ratio method, while simpler, offers a clear snapshot of customer satisfaction. For me I think VADER provides a more comprehensive view

```r
install.packages("sqldf")
library(sqldf)

#Join reviews with products and author_id to get brand name and other columns
reviews_joined <- sqldf("
  SELECT r.review_id, r.product_id, r.review_text, r.rating, r.is_recommended, 
         r.helpfulness, r.total_feedback_count, r.total_neg_feedback_count, 
         r.total_pos_feedback_count, r.submission_date, r.review_title, 
         r.review_length, r.review_type, r.LengthType, r.Pos, r.Neu, 
         r.Neg, r.But, r.Compound, r.rating_category,
         p.brand_name, a.skin_tone, a.eye_color, a.skin_type, a.hair_color
  FROM reviews r
  INNER JOIN products p ON r.product_id = p.product_id
  INNER JOIN authors a ON r.author_id = a.author_id
")
```

```r
unique_brands <- unique(reviews_joined$brand_name)
print(unique_brands)
```

### Calculating VADER scores for both brands

```r
# Calculate the average positive sentiments for L'Occitane and La Mer reviews
loccitane_sentiment <- sqldf("
  SELECT AVG(Pos) AS AvgPos, AVG(Neu) AS AvgNeu, AVG(Neg) AS AvgNeg,
  AVG(But) AS AvgBut, AVG(Compound) AS AvgCompound, STDEV(Compound), COUNT(review_id)
  FROM reviews_joined
  WHERE brand_name = 'L''Occitane'
")
loccitane_sentiment
la_mer_sentiment <- sqldf("
  SELECT AVG(Pos) AS AvgPos, AVG(Neu) AS AvgNeu, AVG(Neg) AS AvgNeg,
  AVG(But) AS AvgBut, AVG(Compound) AS AvgCompound, STDEV(Compound), COUNT(review_id)
  FROM reviews_joined
  WHERE brand_name = 'La Mer'
")
la_mer_sentiment 
```

Our average positive sentiment score stands at 0.1712579, slightly edging out La Mer's score of 0.1690719. This indicates that customers generally express positivity towards our products in their reviews. Moreover, our reviews demonstrate a commendable level of consistency, with a lower standard deviation in compound sentiment (0.415012) compared to La Mer (0.4274761). This suggests that our customers' sentiments are more uniform and less variable, reinforcing the reliability and satisfaction associated with our brand. With 159 reviews analyzed, these findings affirm the strength of our brand's reputation and customer satisfaction levels, providing valuable insights for further enhancing our products and customer experiences.







```r
library(sqldf)
library(tm)
install.packages("wordcloud")
library(wordcloud)
# Filter reviews for L'Occitane and La Mer
loccitane_reviews <- reviews_joined[reviews_joined$brand_name == "L'Occitane", ]

# Clean text data
clean_text <- function(text) {
  text <- tolower(text)
  text <- removePunctuation(text)
  text <- removeNumbers(text)
  text <- removeWords(text, stopwords("en"))
  return(text)
}

# Generate word clouds for L'Occitane
loccitane_high_ratings <- loccitane_reviews %>%
  filter(rating >= 4) %>%
  mutate(clean_review = sapply(review_text, clean_text)) %>%
  pull(clean_review)

loccitane_low_ratings <- loccitane_reviews %>%
  filter(rating <= 2) %>%
  mutate(clean_review = sapply(review_text, clean_text)) %>%
  pull(clean_review)


# Generate word clouds
wordcloud(loccitane_high_ratings, max.words = 100, scale=c(5,0.5))
wordcloud(loccitane_low_ratings, max.words = 100, scale=c(5,0.5))

```

In delving into the reviews of L'Occitane, I've noticed clear distinctions between high and low ratings, offering insights into customer sentiments. High-rated reviews consistently praise specific attributes such as "skin," "cream," and "product," suggesting positive experiences with our skincare range. Expressions like "love," "soft," and "amazing" further underscore customers' satisfaction with the product's effectiveness and quality. Conversely, low-rated reviews often raise concerns about aspects like "smell," "dryness," and "performance," indicating dissatisfaction with fragrance or perceived efficacy. These findings guide our efforts in refining products and addressing customer needs, ultimately improving overall brand perception and customer satisfaction.
